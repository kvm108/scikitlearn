```python
# Logistic reg using 2 features out of the 4
# from IPython.core.display import display, HTML
# https://github.com/rasbt/python-machine-learning-book/blob/master/code/ch03/ch03.ipynb
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets
```

```python
# 1. Load data

iris = datasets.load_iris()
X = iris.data[:, :2] # Only 2 features
Y = iris.target

h = .02 #step size in mesh
```

```python
# 2. Plot data

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5 
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5 
x_min, x_max, y_min, y_max

plt.figure()
plt.clf()

plt.scatter(X[:, 0], X[:, 1], c=Y, cmap= plt.cm.Set1, edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
# 3. Create Model + fit
logreg = linear_model.LogisticRegression(C=1e5)
logreg.fit(X, Y)
```

```python
# 4. Plot decision boundary

# https://stackoverflow.com/questions/36013063/what-is-purpose-of-meshgrid-in-python
# xx,yy - stores each point in the graph between the min max interval. Used to predict all these points inorder to find the decision boundary
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
xx.shape, yy.shape
```

```python
#Convert it into 2 column format
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])
print("{} \n {} \n {}".format(xx.ravel(), yy.ravel(), Z))
print("{} \n {} \n {}".format(xx.ravel().shape, yy.ravel().shape, Z.shape))
```

```python
#-> Stacks 2 rows as a array of columns  
np.c_[np.array([1, 2, 3, 4]), np.array([4, 5, 6, 9])]
```

```python
Z = Z.reshape(xx.shape)
Z
```

```python
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k',
            cmap=plt.cm.Paired)
```
